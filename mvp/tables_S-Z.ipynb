```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
base_url = 'https://www.en.wikipedia.org'

url = 'https://en.wikipedia.org/wiki/List_of_fiction_works_made_into_feature_films_(S%E2%80%93Z)'
page = requests.get(url)
soup = BeautifulSoup(page.text, "lxml")
```

```python
my_tables = soup.find_all('table',{'class':'wikitable'})
```

```python
book_titles = []
movie_titles = []

for table in my_tables:
    rows = table.find_all('tr')
    
    for row in rows:
        cells = row.find_all('td')
        
        
        if len(cells) > 1:
            book_title = cells[0]
            book_titles.append(book_title.text.strip())
            
            movie_title = cells[1]
            movie_titles.append(movie_title.text.strip())
```

```python
df1 = pd.DataFrame({'Book Titles': book_titles,
                    'Movie Titles': movie_titles})
df1
```

```python
df1.to_csv('titles_s_z', sep=',', encoding='utf-8', index=False)
```
